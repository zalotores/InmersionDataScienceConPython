```python
inmuebles['PrecioM2'] = inmuebles['PrecioMillon'] / inmuebles['Area']
inmuebles.sort_values(by='Barrio', ascending=False)
```

```python
f, ax = plt.subplots(figsize=(6, 50))
sns.set_theme(style="whitegrid")
datos = inmuebles.sort_values('Barrio', ascending=False)
sns.barplot(x='PrecioM2', y='Barrio', data=datos, label="Total", color="b", orient='h',
             order=datos.groupby('Barrio')['PrecioM2'].agg('mean').sort_values(ascending=False).index)
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 12), ylabel="", xlabel="Promedio de valor del m2 por barrio")
sns.despine(left=True, bottom=True)
```
