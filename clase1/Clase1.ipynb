```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda'] = valor[0]
inmuebles['Precio'] = valor[1]
inmuebles.sample(3)
```

```python
inmuebles.info()
```

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.', '', regex=True)
inmuebles.sample(4)
```

```python
inmuebles[['Precio', 'Barrio']]
```

```python
inmuebles['PrecioMillon'] = inmuebles.Precio.astype('float') / 1000000
```

```python
inmuebles.info()
```

```python
pd.set_option('display.precision',2)
pd.set_option('display.float.format', lambda x:'%.2f' %x)
```

```python
inmuebles.describe()
```

```python
inmuebles['PrecioMillon'].plot.hist(bins=100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
grafico = sns.histplot(data=inmuebles, x='PrecioMillon', kde=True, hue='Tipo')
grafico.set_title('Distribucion de valores de Inmuebles en Bogota')
plt.xlim((60,1000))
#plt.ylim((0,20))
plt.savefig('/content/drive/MyDrive/Alura/inmersionDataScience/clase1/distribValoresInmuebles.png', format='png')

plt.show()
```

```python
inmuebles['PrecioM2'] = inmuebles['PrecioMillon'] / inmuebles['Area']
inmuebles.sort_values(by='Barrio', ascending=False)
```

```python
barrio = inmuebles.groupby('Barrio')['PrecioM2'].agg(['count','sum'])
barrio['sum'] = barrio['sum'] / barrio['count']
barrio
```

```python
f, ax = plt.subplots(figsize=(6, 50))
sns.set_theme(style="whitegrid")
datos = inmuebles.sort_values('Barrio', ascending=False)
sns.barplot(x='PrecioM2', y='Barrio', data=datos, label="Total", color="b", orient='h',
             order=datos.groupby('Barrio')['PrecioM2'].agg('mean').sort_values(ascending=False).index)
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 10), ylabel="", xlabel="Promedio de valor del m2 por barrio")
sns.despine(left=True, bottom=True)
```
